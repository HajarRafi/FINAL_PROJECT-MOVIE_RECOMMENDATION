```python
#import the libraries
from sqlalchemy import create_engine
import pandas as pd
import getpass
password = getpass.getpass()
```

```python
#create connection with SQL (database 'movie_project' already created)
connection_string = 'mysql+pymysql://root:' + password + '@localhost/movie_project'
engine = create_engine(connection_string)
```

### Read and export to SQL the datasets

```python
movies = pd.read_csv('data/movies_metadata_1.csv')
movies.head()
```

```python
movies.to_sql('movies', con=engine)
```
